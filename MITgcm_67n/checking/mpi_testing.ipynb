```python
# JM: 10 Jun 2020
# checks to compare numbers from some checkpoint output with
# output after changes, for development of GEOMETRIC MPI code

from MITgcmutils import mnc
import matplotlib.pyplot as plt
```

```python
nc = mnc.MNC("../checking_output/mnc_*/GEOMdiag2d.*")
for var in nc.variables:
  print(var)
ene = nc.variables["GEOMeE"][:]
xT  = nc.variables["X"][:]
yT  = nc.variables["Y"][:]
nc.close()

plt.contourf(xT / 1e3, yT / 1e3, ene[-1, 0, :, :], cmap = "RdBu_r", levels = 25)
plt.colorbar()
```

```python
nc = mnc.MNC("../mnc_*/GEOMdiag2d.*")
ene_new = nc.variables["GEOMeE"][:]
xT  = nc.variables["X"][:]
yT  = nc.variables["Y"][:]
nc.close()

plt.contourf(xT / 1e3, yT / 1e3, (ene_new - ene)[-1, 0, :, :], cmap = "RdBu_r", levels = 25)
plt.colorbar()
```
